## Demo: How to run an sql query


```python
# print how many rows are there in the table named 'master'
QUERY = """
SELECT COUNT(*)
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)
df.head()
```

## Demo: How to list all tables


```python
# print all the tables names in the database
QUERY = """
SELECT name as Table_Name FROM
sqlite_master WHERE
type = 'table'
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
pd.read_sql_query(QUERY,conn)
```

## Demo: How to run a group by query


```python
QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
pd.read_sql_query(QUERY,conn)
```

## Demo: How to describe a table


```python
table_name = 'master'  # the table you wish to describe

QUERY = """
SELECT sql FROM sqlite_master
WHERE name= '{}'
""".format(table_name)

df = pd.read_sql_query(QUERY,conn)
print(df.iat[0,0])
```

# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
import numpy as np
import matplotlib.pyplot as plt

query= """
SELECT ConvertedComp 
FROM master
"""
df=pd.read_sql_query(query,conn)

df.plot(kind='hist', bins=10)
plt.title('Histogram of ConvertedComp')
plt.xlabel('ConvertedComp')
plt.ylabel('Count')
plt.show()
```

### Box Plots


Plot a box plot of `Age.`


```python

query="""
SELECT Age
FROM master 
"""
df1=pd.read_sql_query(query,conn)

df1.plot(kind='box',figsize=(10,6))
plt.title('Box Plot of Age')
plt.show()
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
query="""
SELECT Age,WorkWeekHrs
FROM master
"""
df2=pd.read_sql_query(query,conn)

df2.plot(kind='scatter',x='Age',y='WorkWeekHrs',figsize=(10,6))
plt.title('Scatter Plot of Age and WorkWeekHrs')
plt.xlabel('Age')
plt.ylabel('WorkWeekHrs')
plt.show()

```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
import plotly.express as px

query="""
SELECT Age,WorkWeekHrs,CodeRevHrs
FROM master
"""
df3=pd.read_sql_query(query,conn)

df3.plot(kind='scatter',x='WorkWeekHrs',y='CodeRevHrs',s='Age',c='green',figsize=(10,6))
plt.title('Bubble Plot of WorkWeekHrs and CodeRevHrs using Age as size')
plt.xlabel('WorkWeekHrs')
plt.ylabel('CodeRevHrs')
plt.show()


```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
query="""

SELECT count(DatabaseDesireNextYear) as Count, DatabaseDesireNextYear
FROM DatabaseDesireNextYear
GROUP BY DatabaseDesireNextYear
ORDER BY count(DatabaseDesireNextYear) DESC
LIMIT 5

"""

df4=pd.read_sql_query(query,conn)

df4['Count'].plot(kind='pie',labels=df4['DatabaseDesireNextYear'],autopct='%1.1f%%')

plt.show()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
query="""
SELECT WorkWeekHrs,CodeRevHrs,Age
FROM master
WHERE Age BETWEEN 30 AND 35
ORDER BY Age
"""

df5=pd.read_sql_query(query,conn)
df_bar=df5.groupby('Age').median()
df_bar.plot(kind='bar',stacked=True)
plt.title('Median WorkWeekHrs and CodeRevHrs from the age 30 to 35')
plt.xlabel('Age')
plt.ylabel('Hours')
plt.show()

```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
query="""
SELECT ConvertedComp,Age
FROM master
WHERE Age BETWEEN 45 AND 60
"""
df6=pd.read_sql_query(query,conn)
df_line=df6.groupby('Age').median()

df_line.plot(kind='line')

plt.title('Line Chart for Median ConvertedComp from the age of 45 to 60')
plt.xlabel('Age')
plt.ylabel('ConvertedComp')
plt.show()

```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
query="""
SELECT MainBranch, count(MainBranch) as Count
FROM master
GROUP BY MainBranch
"""

df7=pd.read_sql_query(query,conn)

df7.plot(kind='barh')
plt.show()

```
